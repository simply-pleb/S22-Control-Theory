$$10 y^{(4)} -7  y^{(3)} + 2   \ddot y + 0.5 \dot y + 4y  = 15 u$$
Supposing that $y(t)$ and $u(t)$ are functions in terms of $t$.

$$L\left(10 y^{(4)} -7  y^{(3)} + 2   \ddot y + 0.5 \dot y + 4y\right) = L(15 u)$$
$$10L(y^{(4)}) - 7L(y^{(3)}) + 2 L(\ddot y) + 0.5L(\dot y) + 4L(y) = 15 L(u)$$

Supposing that all initial condition are equal to 0 the Laplace transform yields:
$$10s^4Y(s) - 7s^3Y(s) + 2 s^2Y(s) + 0.5sY(s) + 4Y(s) = 15U(s)$$
$$Y(s)\left(10s^4 - 7s^3 + 2 s^2 + 0.5s + 4\right) = 15U(s)$$

$$Y(s) = \frac{15}{10s^4 - 7s^3 + 2 s^2 + 0.5s + 4}U(s)$$

The transfer function is $W(s) = Y(s)/U(s)$, where $U(s)$ is the Laplace transfrom of the input and $Y(s)$ is the Laplace transfrom of the output
$$W(s) = \frac{15}{10s^4 - 7s^3 + 2 s^2 + 0.5s + 4}$$

### Method:
Given a State Space system
$$
\begin{cases}
  \dot x = Ax + Bu \\
  y = Cx
\end{cases}.
$$

Applying Laplace transform to both sides of the first 
equation of the system gives:
- $sIX(s) - AX(s) = BU(s)$;
- $(sI - A)X(s) = BU(s)$;
- $X(s) = (sI - A)^{-1}BU(s)$.

Applying Laplace transform to both sides of the second equation of the system gives:
- $Y(s) = CX(s)$;
- $Y(s) = C(sI - A)^{-1}BU(s)$;
- The transfer function is $W(s) = Y(s)/U(s)$, so
- $W(s) = C(sI - A)^{-1}B$.

In order to find the ODE form of the state space, let us consider the inverse Laplace of the output function:
- let us assume that $W(s)$ is equal to some fraction $N(s)/M(s)$, where $N$ and $M$ are known functions;
- it is known that $W(s) = Y(s)/U(s)$;
- so it can be said that
  $$Y(s)M(s) = U(s)N(s);$$
- let us take the laplace transorm of it
$$\mathcal{L}^{-1}\left(Y(s)M(s)\right)\vert_{t} = \mathcal{L}^{-1}\left(U(s)N(s)\right)\vert_{t};$$
- the ODE would be the result of the inverse Laplace transform.

```python
# Runge-Kutta method
def StateSpace(x, t, A, B):
  # print("not sin(t)")
  return np.dot(A,x)+B

def my_odeint(model, y0, t, A, B):

  t0 = t[0] # get the initial time
  tf = t[-1] # get the final time

  mem = []
  # mem.append(y0)
  
  prev_val = y0
  step_len = (tf-t0)/len(t)
  # mem = np.array(len(t))
  
  for i in range(len(t)):
    temp1 = model(prev_val, t0, A, B)
    temp2 = model(prev_val+step_len/2*temp1, t0+step_len/2, A, B)
    temp3 = model(prev_val+step_len/2*temp2, t0+step_len/2, A, B)
    temp4 = model(prev_val+step_len*temp3,   t0+step_len,   A, B)
    
    cur_step = step_len/6*(temp1+2*temp2+2*temp3+temp4)

    val = prev_val + cur_step
    
    t0 = t0+step_len
    # mem.append(Memory_unit(x, val, lte, gte))
    prev_val = val

    mem.append(val)
  
  mem = np.asarray(mem)

  return mem
    
```

### 3.1.1
 
$\begin{cases}
\ddot x + 0.5 \dot x + 4y = u \\
y = 1.5 \dot x + 6 x
\end{cases}$


Given an ODE
$$\ddot x + 0.5 \dot x + 4(1.5 \dot x + 6 x) = u$$
$$\ddot x + 6.5\dot x + 24x = u.$$

Since the initial condition are zero, the Laplace transform would be the following
$$s^2X(s) + 6.5sX(s) + 24X(s) = U(s)$$
$$X(s) = \frac{U(s)}{s^2 + 6.5s + 24}.$$

Since $$Y(s)=(1.5s + 6)X(s)$$ 

$$Y(s)=\frac{(1.5s + 6)U(s)}{s^2 + 6.5s + 24},$$ 

and the transfer function is $W(s) = Y(s)/U(s)$, so
$$W(s) = \frac{1.5s + 6}{s^2 + 6.5s + 24}.$$

### 3.1.2

$$10 \ddot x + 1.5 \dot x + 8(15 \dot x + 16 x) = 0.5u$$
$$10 \ddot x + 121.5\dot x + 148 x = 0.5u$$

Since the initial condition are zero, the Laplace transform would be the following
$$10s^2X(s) + 121.5sX(s) + 128X(s) = 0.5U(s)$$
$$X(s) = \frac{0.5U(s)}{10s^2 + 121.5s + 128}.$$

Since $$Y(s)=(15s + 16)X(s)$$ 

$$Y(s)=\frac{(15s + 16)U(s)}{20s^2 + 243s + 256},$$ 

and the transfer function is $W(s) = Y(s)/U(s)$, so
$$W(s) = \frac{15s + 16}{20s^2 + 243s + 256}.$$

# 4. Stability of an autonomous linear system

Autonomous linear system is *stable*, iff the eigenvalues of its matrix have negative real parts. In other words, their should lie on the left half of the complex plane.

Consider the system:

$$\dot x = 
\begin{pmatrix} -1 & 0.4 \\ -20 & -16
\end{pmatrix}
x
$$

Let us find its eigenvalues:

$$Ax = \lambda x$$
$$(A-Iλ)x = 0$$
$$det(A-Iλ) = 0$$

- Finding the eigenvalues of the given equation yeilds:

$ \begin{vmatrix} -1-\lambda & 0.4 \\ -20 & -16-\lambda \end{vmatrix} = 0 $\
$$(1+\lambda)(16+\lambda) + 8 = 0$$
$$24+\lambda^2+17\lambda = 0$$
$$\lambda_{1,2}=\frac{-17\pm\sqrt{17^2-24\cdot 4}}{2}$$
$$\lambda_{1,2}=\frac{-17\pm 13.89}{2}$$

- The eigenvalues of the given system are:
$$\lambda_1=-1.55,\ \ \lambda_2=-15.44$$

Since all eigenvalues are negative, the given autonomous linear system is stable. 


```python
A = np.array([[   -1,    0.4 ],
              [  -20,   -16]])

Lambda, Q = np.linalg.eig(A)
print(f"Eigen values:\n{Lambda}, \n\n Eigen vectors:\n{Q}\n\n")

B = np.array([0, 0])

# x0 = np.zeros(len(B))
x0 = np.random.rand(len(B))  # random initial state

simulate_ss(StateSpace, x0, A, B)
```
